# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(independent, dependent)
```

```python
lin_reg.predict([[0.0, 0.0, 1.0, 0, 1, 85]])
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree=4)
```

```python
X_poly = poly_reg.fit_transform(independent)
```

```python
lin_reg2 = LinearRegression()
```

```python
lin_reg2.fit(X_poly, dependent)
```

```python
lin_reg2.predict(poly_reg.fit_transform([[0.0, 0.0, 1.0, 0, 1, 85]]))
```

# SVR

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_scaler = scaler.fit_transform(independent)
y_scaler = scaler.fit_transform(dependent)
```

```python
from sklearn.svm import SVR
```

```python
sv_reg = SVR(kernel='rbf')
```

```python
sv_reg.fit(x_scaler, y_scaler)
```

```python
sv_reg.predict([[0.0, 0.0, 1.0, 0, 1, 85]])
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt_reg = DecisionTreeRegressor(random_state=0)
```

```python
dt_reg.fit(independent, dependent)
```

```python
dt_reg.predict([[0.0, 0.0, 1.0, 0, 1, 85]])
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_reg = RandomForestRegressor(n_estimators=10, random_state=0)
```

```python
rf_reg.fit(independent, dependent)
```

```python
rf_reg.predict([[0.0, 0.0, 1.0, 0, 1, 85]])
```
