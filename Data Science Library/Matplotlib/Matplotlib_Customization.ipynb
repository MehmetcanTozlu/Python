```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

## 1'den 10'a kadar Esit Araliklarla 20 tane veri uret;

```python
numpyArray = np.linspace(1, 10, 20)
```

```python
numpyArray
```

### numpyArray Array'inin de 3. kuvvetini alip bir degiskene atayalim ve grafige dokelim;

```python
numpyArray2 = numpyArray ** 3
```

```python
numpyArray2
```

```python
plt.plot(numpyArray, numpyArray2, 'g--')
```

## Simdi de 1 grafik uzerinde 2 durumuda gerceklestirmeye calisalim;

plt.subplot(row'daki grafik sayisi, column'daki grafik sayisi, cizilecek grafigin index sayisi)

```python
plt.subplot(1, 2, 1) # 1 satirdan, 2 sutundan olusacak ve 1. grafigi cizicem
plt.plot(numpyArray, numpyArray2, 'ko-')
plt.subplot(1, 2, 2) # 1 satirdan, 2 sutundan olusacak ve 2. grafigi cizicem
plt.plot(numpyArray2, numpyArray, 'r*-')
```

### Ustteki grafigi yan yana yerine alt alta cizdirelim;

```python
plt.subplot(2, 1, 1) # 2 satir, 1 sutun ve 1. indexi cizicem
plt.plot(numpyArray, numpyArray2, 'ko-')
plt.subplot(2, 1, 2) # 2 satir, 1 sutun ve 2. indexi cizicem
plt.plot(numpyArray2, numpyArray, 'r*-')
```

# Figure;

### alttaki kullanim matplotlib class'inin figure ozelliginden bir nesne urettik
### bu sekilde nesne uretip kullanmak bize daha ozgur bir kullanim saglamaktadir.

figur.add_axes([y-eks.'nin baslingic noktasi, x-eks'nin baslangic noktasi, y-eks.'nin uzunlugu, x-eks'nin uzunlugu])  # **args veya **kwargs alir o yuzden list biciminde verdik

```python
figur = plt.figure()
figurAxes = figur.add_axes([0.2, 0.2, 1.0, 1.0])
figurAxes.plot(numpyArray, numpyArray2, 'b')
figurAxes.set_xlabel('X-Ekseni')
figurAxes.set_ylabel('Y-Ekseni')
figurAxes.set_title('Grafik Basligi')
```
