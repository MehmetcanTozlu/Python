# Gorsel Gelistirmeler;

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
numpyArray = np.linspace(1, 10, 20)
```

```python
numpyArray2 = numpyArray ** 3
```

## Ayni grafik uzerinde 1'den fazla eksen gostermek istersek;

plt.figure()'e dpi degeri verebiliriz, dip degeri yuksek olursa grafik buyuyerek cozunurlugunu arttirabilecegimiz bir ozelliktir

```python
figur = plt.figure(dpi=100)

axes1 = figur.add_axes([0.1, 0.1, 1.3, 1.3])
axes1.plot(numpyArray, numpyArray2, label='numpyArray ** 3')
axes1.plot(numpyArray, numpyArray ** 2, label='numpyArray ** 2')
axes1.plot(numpyArray, numpyArray ** 3.3, label='numpyArray ** 3.3')
axes1.legend(loc=0) # label etiketini legend() fonksiyonu ile yazdiriyoruz 1 parametre alir loc
```

# Belki cizdirdigimiz grafigi kaydetmek isteyebiliriz;
