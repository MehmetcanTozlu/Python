# Ileri Seviye Gorsel Gelistirmeler

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
numpyArray = np.linspace(1, 10, 20)
```

```python
numpyArray
```

```python
numpyArray2 = numpyArray ** 2
```

```python
numpyArray2
```

### Dogru renklerini hexadecimal olarak ayarlayalim;

plot() fonksiyonu alpha degeride alabilir.
alpha=1'den alpha=0 araliginda deger alir. Amaci transparent'dir. 1 koyu, 0 seffaf!

```python
(figur, axes) = plt.subplots(nrows=1, ncols=1)
axes.plot(numpyArray, numpyArray2, color='#FF56A3')
axes.plot(numpyArray2, numpyArray, color='#00F2FF', alpha=0.3) # rgb renklerin hex-code olarak da renk verebiliriz
```

## Dogrunun gorunurluklerini degistirmek istersek;

```python
(figur2, axes2) = plt.subplots(dpi=150)

# dogrunun kalinliklarini degistirdik
axes2.plot(numpyArray, numpyArray + 2, color='indigo', linewidth=4.0)
axes2.plot(numpyArray, numpyArray + 3, color='red', linewidth=1.0)

# dogrunun gorunur seklini degistirdik
axes2.plot(numpyArray, numpyArray + 4, color='blue', linestyle=':')
axes2.plot(numpyArray, numpyArray + 5, color='green', linestyle='-.')

# dogru uzerinde x ve y eksenlerinin kesistikleri yerleri belirttik
axes2.plot(numpyArray, numpyArray, color='#066700', linestyle='--', marker='+')

# dogrunun x ve y eksenlerinde kesistikleri yerleri o ile gosterdik ve iclerini siyaha boyattik
axes2.plot(numpyArray, numpyArray - 1, color='#066700', linestyle='--', marker='o', markersize=7, markerfacecolor='k')
```
