## Ic Ice grafik yapmak istersek;

add_axes() fonksiyonun icindeki axes1 -> 0.1, 0.1 veya axes2 -> 0.3, 0.6 2 grafigin birbirlerine olan konumlarini ayarlamamiza yardimci oluyor. Ilk girilen y eks.'e olan yakinligi gosteriyor ne kadar kucuk olursa y eks'e o kadar yakin oluyorlar,
ikinci girilende x eks'i gosteriyor ne kadar kucuk olursa 2 grafikte o kadar x eks'e yakin oluyor!

```python
figur = plt.figure()
axes1 = figur.add_axes([0.1, 0.1, 1.0, 1.0]) # 1. yarattigimiz nesne alttaki    ilk girilen 2 deger birbirlerine olan
axes2 = figur.add_axes([0.3, 0.6, 0.3, 0.3]) # 2. yarattigimiz nesne ustteki    konumlarini belirliyor!

axes1.set_xlabel('Axes-1 X Axes')
axes1.set_ylabel('Axes-1 Y Axes')
axes1.set_title('Axes-1 Title')
axes1.plot(numpyArray1, numpyArray2, 'r')

axes2.set_xlabel('Axes-2 X Axes')
axes2.set_ylabel('Axes-2 Y Axes')
axes2.set_title('Axes-2 Title')
axes2.plot(numpyArray2, numpyArray1, 'k')
```
